```
import numpy as np
np.random.seed(5980202)

n= 20
m= 100
A = np.zeros((m,n))
temp = np.random.uniform(size=n)
for i in range(m):
  A[i,:] = temp
x = np.random.randn(n,1)
y = np.zeros((m,1))
for i in range(m):
  y[i,:] = np.dot(A[i,:],x)*np.dot(A[i,:],x)
```

```
loss_f = lambda A1,x1,y1 :  1/4*np.dot((y1-np.multiply(np.dot(A1,x1),np.dot(A1,x1))).T,y-np.multiply(np.dot(A1,x1),np.dot(A1,x1)))
```

```
x = np.random.randn(n,1)
eps = 1e-30
iteration = 0
while True:
  iteration += 1
  loss = loss_f(A,x,y)
  if loss[0][0] < eps:
    break
  res = np.zeros((m,1))
  for i in range(res.shape[0]):
    res[i,:] = y[i,:]-np.dot(A[i,:],x)*np.dot(A[i,:],x)
  for i in range(res.shape[0]):
    ai = A[i].reshape(A[i].shape[0],1)
    if i==0:
      grad = -res[i]*np.dot(x.T,ai)*ai
    else:
      grad += -res[i]*np.dot(x.T,ai)*ai
  grad = grad.reshape(grad.shape[0],1)
  for i in range(res.shape[0]):
    res[i,:] = y[i,:]-3*np.dot(A[i,:],x)*np.dot(A[i,:],x)
  for i in range(res.shape[0]):
    ai = A[i].reshape(A[i].shape[0],1)
    if i==0:
      hessian = -res[i]*np.dot(ai,ai.T)
    else:
      hessian += -res[i]*np.dot(ai,ai.T)
  x = x-np.dot(np.linalg.pinv(hessian),grad)
  print('iteration',iteration,' : loss',loss[0][0])

print("loss =",loss[0][0])
```
